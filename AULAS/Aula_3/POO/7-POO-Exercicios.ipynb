```python
# Exercício 1 - Crie um objeto a partir da classe abaixo, chamado roc1, 
# passando 2 parâmetros e depois faça uma chamada aos atributos e métodos
from math import sqrt
from random import randint

class Rocket():
     
    def __init__(self, x=0, y=0, cont=0):
        self.x = x
        self.y = y
        self.cont = cont

    def move_rocket(self, x_increment=0, y_increment=1):
        self.x += x_increment
        self.y += y_increment
        self.cont += 1

    def print_rocket(self):
        print(self.x, self.y)

    def print_contador(self):
        print(f"Contador de movimentos: {self.cont}")

    def randomize_position(self):
        self.x = randint(0, 100)
        self.y = randint(0, 100)

# Crie o objeto roc1 passando 2 parâmetros
roc1 = Rocket(10, 20)

# Chame os métodos e atributos
roc1.print_rocket()         
roc1.move_rocket(5, 7)      
roc1.print_rocket()         
roc1.print_contador()       

roc1.randomize_position()   
roc1.print_rocket()        
```
